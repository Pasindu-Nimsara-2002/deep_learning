```python
import cv2
import numpy as np

def enhance_and_detect_features(image_path):
    # Load the thermal image (grayscale)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Error: Image not found.")
        return

    # Convert the grayscale image to BGR (for easier manipulation)
    image_bgr = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    # Convert the BGR image to HSV
    image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

    # Extract the V (brightness) channel
    h, s, v = cv2.split(image_hsv)

    # Step 1: Apply CLAHE for contrast enhancement on the V channel
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    v = clahe.apply(v)

    # Step 2: Apply Gaussian Blur to reduce noise
    blurred_image = cv2.GaussianBlur(v, (5, 5), 0)

    # Step 3: Apply Adaptive Thresholding to create a binary image
    thresh_image = cv2.adaptiveThreshold(blurred_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

    # Step 4: Edge detection using Canny
    edges = cv2.Canny(thresh_image, threshold1=50, threshold2=150)

    # Step 5: Merge the processed V channel back with H and S channels
    enhanced_image_hsv = cv2.merge([h, s, v])

    # Step 6: Convert back to BGR for display
    enhanced_image_bgr = cv2.cvtColor(enhanced_image_hsv, cv2.COLOR_HSV2BGR)

    # Resize images to fit within a 640x480 window
    resized_image = cv2.resize(image, (640, 480))
    resized_enhanced_image = cv2.resize(enhanced_image_bgr, (640, 480))
    resized_thresh_image = cv2.resize(thresh_image, (640, 480))
    resized_edges = cv2.resize(edges, (640, 480))

    # Display the resized images
    cv2.imshow('Original Image', resized_image)
    cv2.imshow('Enhanced Image', resized_enhanced_image)
    cv2.imshow('Thresholded Image', resized_thresh_image)
    cv2.imshow('Edge Detected Image', resized_edges)

    # Wait for the user to press a key
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Call the function with your image path
image_path = 'image.png'  # Replace with your image path
enhance_and_detect_features(image_path)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Define custom LUT transformation points
c = np.array([(50, 100), (200, 150)])

# Generate LUT arrays using np.linspace
t1 = np.linspace(0, c[0, 1], c[0, 0] + 1, endpoint=True).astype('uint8')
t2 = np.linspace(c[0, 1] + 1, c[1, 1], c[1, 0] - c[0, 0], endpoint=True).astype('uint8')
t3 = np.linspace(c[1, 1] + 1, 255, 255 - c[1, 0], endpoint=True).astype('uint8')

# Concatenate arrays to form the LUT
transform = np.concatenate((t1, t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')

# Ensure the LUT has 256 values (from 0 to 255)
transform = transform[:256]  # To ensure the LUT is 256 elements long

# Load the image
img_orig = cv.imread('image.png', cv.IMREAD_GRAYSCALE)

# Apply the custom LUT transformation
image_transformed = cv.LUT(img_orig, transform)

# Display the original and transformed images
plt.subplot(1, 2, 1)
plt.imshow(img_orig, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(image_transformed, cmap='gray')
plt.title('Transformed Image')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)
blurred = cv2.GaussianBlur(image, (5,5), 0)

# Apply Sobel filters
sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=5)  # X-direction
sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=5)  # Y-direction

# Convert to absolute values and normalize
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)

# Combine both gradients
sobel_combined = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

# Display results
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel X')
plt.subplot(1, 3, 2), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel Y')
plt.subplot(1, 3, 3), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Combined')
plt.show()
```

```python
import cv2
import numpy as np

def unsharp_mask(image, sigma=1.0, strength=1.5):
    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(image, (0, 0), sigma)
    # Subtract the blurred image from the original
    sharpened = cv2.addWeighted(image, 1.0 + strength, blurred, -strength, 0)
    return sharpened

# Load the image
image = cv2.imread('image.png')
# Apply unsharp masking
a_sharpened_image = unsharp_mask(image)
# Save the result


# Laplacian Filter with OpenCV
import cv2
import numpy as np

def unsharp_mask_Laplacian(image, sigma=1.0, strength=1.5, kernel_size=(5, 5)):
    # Apply Gaussian blur with specified kernel size
    blurred = cv2.GaussianBlur(image, kernel_size, sigma)
    # Subtract the blurred image from the original
    sharpened = cv2.addWeighted(image, 1.0 + strength, blurred, -strength, 0)
    return sharpened

# Load the image
image = cv2.imread('image.png')
# Apply unsharp masking with adjusted parameters
b_sharpened_image = unsharp_mask_Laplacian(image, sigma=3.0, strength=5.5, kernel_size=(5, 5))
# Save the result


#High-Pass Filter with OpenCV
import cv2
import numpy as np

def high_pass_filter(image, sigma=1.0):
    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(image, (0, 0), sigma)
    # Subtract the blurred image from the original
    high_pass = cv2.subtract(image, blurred)
    # Add the high-pass image back to the original
    sharpened = cv2.addWeighted(image, 1.0, high_pass, 1.0, 0)
    return sharpened

# Load the image
image = cv2.imread('image.png')
# Apply high-pass filter
c_sharpened_image = high_pass_filter(image)
# Save the result


# Display results
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1), plt.imshow(a_sharpened_image, cmap='gray'), plt.title('Sobel X')
plt.subplot(1, 3, 2), plt.imshow(b_sharpened_image, cmap='gray'), plt.title('Sobel Y')
plt.subplot(1, 3, 3), plt.imshow(c_sharpened_image, cmap='gray'), plt.title('Sobel Combined')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

# Sobel X and Y
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

# Compute gradient at 45-degree angle
theta = np.deg2rad(45)  # Convert 45 degrees to radians
sobel_45 = np.cos(theta) * sobel_x + np.sin(theta) * sobel_y

# Convert to absolute
sobel_45 = cv2.convertScaleAbs(sobel_45)

# Show the result
plt.imshow(sobel_45, cmap='gray')
plt.title('Sobel at 45 Degrees')
plt.show()
```

```python
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

scharr_x = cv2.Scharr(image, cv2.CV_64F, 1, 0)  # Detects vertical-like edges
scharr_y = cv2.Scharr(image, cv2.CV_64F, 0, 1)  # Detects horizontal-like edges

scharr_combined = cv2.addWeighted(cv2.convertScaleAbs(scharr_x), 0.5, 
                                  cv2.convertScaleAbs(scharr_y), 0.5, 0)

plt.imshow(scharr_combined, cmap='gray')
plt.title('Scharr Edge Detection')
plt.show()
```
