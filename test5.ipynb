```python
import cv2
import numpy as np

# Load an image
image = cv2.imread("image.png")  # Change the path to your image
image = cv2.resize(image, (600, 400))  # Resize for display
original = image.copy()

# Trackbar callback function
def update_image(_):
    brightness = cv2.getTrackbarPos("Brightness", "Adjustments") - 100
    contrast = cv2.getTrackbarPos("Contrast", "Adjustments") / 50.0
    exposure = cv2.getTrackbarPos("Exposure", "Adjustments") / 50.0
    highlight = cv2.getTrackbarPos("Highlights", "Adjustments") / 50.0

    # Adjust Brightness & Contrast
    adjusted = cv2.convertScaleAbs(original, alpha=contrast, beta=brightness)

    # Adjust Exposure
    adjusted = cv2.convertScaleAbs(adjusted, alpha=exposure, beta=0)

    # Adjust Highlights (emphasizing bright regions)
    hsv = cv2.cvtColor(adjusted, cv2.COLOR_BGR2HSV)
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * highlight, 0, 255)
    adjusted = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    cv2.imshow("Adjustments", adjusted)

# Create a window
cv2.namedWindow("Adjustments")

# Create trackbars
cv2.createTrackbar("Brightness", "Adjustments", 100, 200, update_image)
cv2.createTrackbar("Contrast", "Adjustments", 50, 100, update_image)
cv2.createTrackbar("Exposure", "Adjustments", 50, 100, update_image)
cv2.createTrackbar("Highlights", "Adjustments", 50, 100, update_image)

# Initial call to update the image
update_image(0)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
