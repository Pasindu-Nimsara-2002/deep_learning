```python
import cv2

# Load the grayscale thermal image
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

# Apply noise reduction
denoised_image = cv2.medianBlur(image, 5)

# Apply contrast stretching
stretched_image = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX)

# Apply pseudocoloring
pseudocolored_image = cv2.applyColorMap(stretched_image, cv2.COLORMAP_HOT)

# Save or display the result
cv2.imwrite('enhanced_thermal_image.png', pseudocolored_image)
cv2.imshow('Enhanced Image', pseudocolored_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale thermal image
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)

# Step 1: Noise Reduction (Median Blur)
denoised_image = cv2.medianBlur(image, 5)

# Step 2: Contrast Stretching
min_val = np.min(denoised_image)
max_val = np.max(denoised_image)
stretched_image = cv2.normalize(denoised_image, None, 0, 255, cv2.NORM_MINMAX)

# Step 3: Histogram Equalization (CLAHE for adaptive equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
equalized_image = clahe.apply(stretched_image)

# Step 4: Sharpening (Using a sharpening kernel)
sharpening_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_image = cv2.filter2D(equalized_image, -1, sharpening_kernel)

# Step 5: Pseudocoloring (Apply a colormap for better visualization)
pseudocolored_image = cv2.applyColorMap(sharpened_image, cv2.COLORMAP_JET)

# Convert BGR (OpenCV default) to RGB for Matplotlib
pseudocolored_image_rgb = cv2.cvtColor(pseudocolored_image, cv2.COLOR_BGR2RGB)

# Display the results using Matplotlib
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Thermal Image')
plt.axis('off')

# Enhanced Image
plt.subplot(1, 2, 2)
plt.imshow(pseudocolored_image_rgb)
plt.title('Enhanced Thermal Image (Pseudocolored)')
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
